```python
import pandas as pd

# 예시 데이터프레임 생성
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

# 각 열의 합 계산 (axis=0)
sum_by_column = df.sum(axis=0)
print("Sum by column (axis=0):\n", sum_by_column)

# 각 행의 합 계산 (axis=1)
sum_by_row = df.sum(axis=1)
print("\nSum by row (axis=1):\n", sum_by_row)

df
```
