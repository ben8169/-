```python
import pandas as pd
import numpy as np
from SVM import *
```

```python
train_raw = pd.read_csv('preprocessed_train.csv')
train = train_raw.iloc[:,1:]
train
```

```python
remaining_cols = train.columns.drop('Country.of.Origin')
X = train[remaining_cols]
Y = train['Country.of.Origin']
X = X.fillna(0)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
X_train.shape, Y_train.shape
```

```python
model = MulticlassSVM()
X_train_np = X_train.to_numpy().astype('float64')
Y_train_np = Y_train.astype('float64')
X_test_np = X_test.to_numpy().astype('float64')
Y_test_np = Y_test.to_numpy().astype('float64')
```

```python
# Accuracy doesn't improve well. Small lr, lambda_params are effective.
# hyperparameters
n_iters_candidates = [5000,10000,15000,20000,30000]
lr_candidates = [0.0001, 0.001, 0.01, 0.1]
lambda_param_candidates = [0.001, 0.01, 0.1, 1, 10]

best_accuracy = 0
best_params = {}

cnt = 0
for n_iters in n_iters_candidates:
    for lr in lr_candidates:
        for lambda_param in lambda_param_candidates:
            model = MulticlassSVM(n_iters=n_iters, lr=lr, lambda_param=lambda_param)
            model.fit(X_train_np, Y_train_np)
            
            y_pred = model.predict(X_test_np)
            accuracy = model.get_accuracy(Y_test_np, y_pred)
            
            if accuracy > best_accuracy:
                best_accuracy = accuracy
                best_params = {'n_iters': n_iters, 'lr': lr, 'lambda_param': lambda_param}
            cnt += 1
            print(f"{cnt}st: accuracy:{accuracy}, 'n_iters': {n_iters}, 'lr': {lr}, 'lambda_param': {lambda_param}")

print(f"Best Accuracy: {best_accuracy:.2f}")
print(f"Best Parameters: {best_params}")
```

#### performance metric

```python
def accuracy(y_true, y_pred):
    correct = 0
    for yt, yp in zip(y_true, y_pred):
        if yt == yp:
            correct += 1
    return correct / len(y_true)

def precision(y_true, y_pred):
    true_positives = 0
    predicted_positives = 0
    for yt, yp in zip(y_true, y_pred):
        if yt == 1:
            if yp == 1:
                true_positives += 1
        if yp == 1:
            predicted_positives += 1
    if predicted_positives == 0:
        return 0
    return true_positives / predicted_positives

def recall(y_true, y_pred):
    true_positives = 0
    actual_positives = 0
    for yt, yp in zip(y_true, y_pred):
        if yt == 1:
            actual_positives += 1
            if yp == 1:
                true_positives += 1
    if actual_positives == 0:
        return 0
    return true_positives / actual_positives

def f1_score(y_true, y_pred):
    p = precision(y_true, y_pred)
    r = recall(y_true, y_pred)
    if p + r == 0:
        return 0
    return 2 * (p * r) / (p + r)

print(accuracy(Y_test, Y_pred))
print(precision(Y_test, Y_pred))
print(recall(Y_test, Y_pred))
print(f1_score(Y_test, Y_pred))
```
