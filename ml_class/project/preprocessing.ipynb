### Preprocessing

#### Altitude

```python
train_drop.dropna()
```

```python
import re

altitudes = train_drop['Altitude']

def preprocess_altitudes(altitudes):
    processed_altitudes = []
    cnt = 0
    for altitude in altitudes:
        if cnt == 298: print(altitude)
        # Remove alphabets and special characters
        cleaned_altitude = re.sub(r'[^0-9\-]+', '', altitude)

        if cleaned_altitude == '190164': continue
        # Handle ranges
        if '-' in cleaned_altitude:
            start, end = map(int, cleaned_altitude.split('-'))
            average = (start + end) / 2
            processed_altitudes.append(average)
        elif len(cleaned_altitude)>=7:
            start, end = int(cleaned_altitude)/10000, int(cleaned_altitude)%10000
            average = (start + end) / 2
            processed_altitudes.append(average)
        else:
            processed_altitudes.append(int(cleaned_altitude))
        cnt += 1
    
    return processed_altitudes

preprocessed_altitudes = preprocess_altitudes(altitudes)


```
