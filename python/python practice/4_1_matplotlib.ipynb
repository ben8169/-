### Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.tri import (
        Triangulation, UniformTriRefiner, CubicTriInterpolator)
```

```python
print(plt.style)
[print(i) for i in plt.style.available]

x = np.arange(10)
y = x + np.random.randint(-10, 10, len(x))

plt.style.context('_mpl-gallery')
plt.subplot(1, 2, 1)
plt.plot(x,y)

plt.subplot(1, 2, 2)
plt.bar(x,y)
plt.show()
```

```python
plt.figure(figsize=(12,10))
for i, style in enumerate(plt.style.available):
    with plt.style.context(style):
        plt.subplot(5, 6, i+1)
        plt.bar(x,y)

plt.tight_layout()
plt.show()
# plt.savefig('./plot_2.png')
```

```python
np.random.seed(21)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

colors = ['r', 'g', 'b', 'k']
yticks = np.linspace(0, 1.0, 4)

for c, k in zip(colors, yticks):
    xs = np.arange(20)
    ys = np.random.rand(20)

    cs = [c] * len(xs)
    cs[0] = 'c'
    ax.bar(xs, ys, zs=k, zdir='y', color=cs, alpha=0.8)


ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_yticks(yticks)
plt.show()
```

### Seaborn

```python
import seaborn as sns

sns.set_theme(style="whitegrid")

tips = sns.load_dataset("tips")


sns.violinplot(data=tips, x="day", y="total_bill", hue="smoker",
                split=True, inner="quart", linewidth=1,
                palette={"Yes": "b", "No": ".85"})

sns.despine(left=True)
plt.show()
```

### mpld3

```python
import mpld3
import pandas as pd

def plot_1():
    def sample_plot(df, title):
        df.plot(kind='line',
                marker='p',
                color=['blue', 'red'],
                lw=3,
                ms=20,
                alpha=0.7)
        plt.title(title)
        plt.text(s='blue line', x=1.7, y=2.5, color='blue')
        plt.text(s='red line', x=2.6, y=1.3, color='red')

    data = {
        'c1': [1, 3, 2, 4],
        'c2': [3, 4, 1, 2]
    }
    df = pd.DataFrame(data)

    sample_plot(df, 'mpld3')
    mpld3.show()
    mpld3.display()
plot_1()
```

```python
sns.set_theme(style="whitegrid")

tips = sns.load_dataset("tips")

sns.violinplot(data=tips, x="day", y="total_bill", hue="smoker",
                   split=True, inner="quart", linewidth=1,
                   palette={"Yes": "b", "No": ".85"})
sns.despine(left=True)
mpld3.show()
```
